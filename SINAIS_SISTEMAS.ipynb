# Decomposição de sinais

Implemente uma função que decomponha um sinal de tempo discreto $x[n]$ em suas partes par $x_p[n]$ e impar
$x_i[n]$.

• Entrada:

– Sinal a ser decomposto $x[n]$ (arranjo/vetor com os valores da variável dependente).

– Valor do tempo inicial de referência $n_0$ (variável independente).

• Saída:

– Sinal par $x_p[n]$ (arranjo/vetor com valores da variável dependente).

– Sinal impar $x_i[n]$ (arranjo/vetor com valores da variável dependente).
– Valor do tempo inicial $n_0$ (variável independente).

```python
alpha=0.5
A=1
n1 = np.linspace(-4,4)
x1 = A*alpha**n1

plt.stem(n1, x1, use_line_collection=True)
plt.xlabel('n')
plt.ylabel('$x_1[n]$')
plt.show()
```

```python
n2 = np.linspace(-10,10 )
omega=2*np.pi*10
x2 = np.sin(omega*n2)

plt.stem(n2 ,x2, use_line_collection=True)
plt.xlabel('n')
plt.ylabel('$x_2[n]$')
plt.show()
```

```python
def signaldecomposition(xmin,xma):
  xp=(xma+xmin)/2
  xi=(xma-xmin)/2
  return xp,xi
```

```python
x1new=n1*x1
plt.stem(n1 ,x1new, use_line_collection=True)
plt.xlabel('n')
plt.ylabel('$nx_1[n]$')
plt.show()
```

```python
plt.stem(n1 ,signaldecomposition(-n1*A*alpha**(-n1),n1*A*alpha**(n1))[0], use_line_collection=True)
plt.xlabel('n')
plt.ylabel('$nx_1p[n]$')
plt.title('Parte par de $nx_1i[n]$')
plt.show()
```

```python
plt.stem(n1 ,signaldecomposition(-n1*A*alpha**(-n1),n1*A*alpha**(n1))[1], use_line_collection=True)
plt.xlabel('n')
plt.ylabel('$nx_1i[n]$')
plt.title('Parte ímpar de $nx_1i[n]$')
plt.show()
```

```python
oddpartofx1 =signaldecomposition(-n1*A*alpha**(-n1),n1*A*alpha**(n1))[1]
evenpartofx1=signaldecomposition(-n1*A*alpha**(-n1),n1*A*alpha**(n1))[0]
x1sum=oddpartofx1+evenpartofx1
plt.stem(n1 ,x1sum, use_line_collection=True)
plt.xlabel('n')
plt.ylabel('$nx_1[n]$')
plt.title('Soma das partes par e ímpar de $nx_1[n]$')
plt.show()
```

O sinal obitdo coincide com o sinal original que um sinal pode ser decomposto da seguinte forma:

$$
x[n] =x_p[n] +x_i[n]
$$

Onde:
$$
x_p\left ( t \right ) = \frac{x[n]+  x[-n]}{2}
$$
e
$$
x_i[n]= \frac{xi[n]- xi[-n]}{2}
$$

# Soma de convolução

Implemente uma função que realize a convolução entre dois sinais de tempo discreto $x[n]$ e $h[n]$. A função
deve apresentar a seguinte interface:

• Entrada:

– Sinal $x[n]$ (arranjo/vetor com os valores da variável dependente).

– Sinal $h[n]$ (arranjo/vetor com os valores da variável dependente).

– Valor do tempo inicial de referência $n 0$ (variável independente).
• Saída:
– Sinal par $y[n]$ (arranjo/vetor com valores da variável dependente).

– Valor do tempo inicial $n_0$ (variável independente).

Considerando $h[n]$ como sendo um degrau unitário:
$$
h[n]=u[n]
$$

```python
def convolution(x,h):
  m=len(x)
  n=len(h)
  y=np.zeros(n+m+1)
  for  i in range(0,n+m):
    y[i]=0
    for j in range(0,m):
      if(i-j+1>0 and i-j+1<50  ):
      
        y[i]=y[i]+x[j]*h[i-j+1]; 
  return y
```

```python
h=np.heaviside(n1, 1) 
plt.stem(n1 ,h, use_line_collection=True)
plt.xlabel('n')
plt.ylabel('$h[n]$')
plt.show()
```

```python
x1new=n1*x1
plt.stem(n1 ,x1new, use_line_collection=True)
plt.xlabel('n')
plt.ylabel('$nx_1[n]$')
plt.show()
```

```python
plt.stem(n1 ,n1*h, use_line_collection=True)
plt.xlabel('n')
plt.ylabel('$nh[n]$')
plt.show()
```

```python
xh = convolution(x1,h)
nnew=np.linspace(min(n1),max(n1),len(xh))
plt.stem(nnew,xh, use_line_collection=True)
plt.xlabel('n')
plt.ylabel('$x_1[n]*h[n]$')
 
```

```python
xh = convolution(x1,h)
nnew=np.linspace(min(n1),max(n1),len(xh))
plt.stem(nnew,nnew*xh, use_line_collection=True)
plt.xlabel('n')
plt.ylabel('$y=n(x_1[n]*h[n])$')
 
```

```python
xh = convolution(x1new,n1*h)
nnew=np.linspace(min(n1),max(n1),len(xh))
plt.stem(nnew,xh, use_line_collection=True)
plt.xlabel('n')
plt.ylabel('$nx_1[n]*nh[n]$')
```

```python
xh = convolution(x1new,n*h)
nnew=np.linspace(min(n1),max(n1),len(xh))
plt.stem(nnew,nnew*xh, use_line_collection=True)
plt.xlabel('n')
plt.ylabel('$y=n(x_1[n]*h[n])$')
 
```
